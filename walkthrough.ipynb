They come in 2 sizes. We are only intrested in data ranges.

```python
RANGE_RECT = {
  (640, 480): {
    "left": 546,
    "right": 576,
    "top": 82,
    "bottom": 100
  },
  (672, 512): {
    "left": 579,
    "right": 612,
    "top": 82,
    "bottom": 100
  }
}
```

```python
images = [image1, image2, image3]

def get_crop_rect(image):
    rect = RANGE_RECT[image.size]
    left = rect['left']
    top = rect['top']
    right = rect['right'] 
    bottom = rect['bottom']
    return (left, top, right, bottom)
    
    
image1.crop(get_crop_rect(image1))
```

We load a sample image containing all digits:

```python
sample = Image.open('sample.bmp')
sample
```

If we are lucky, vertical histogram alone should be enough to distigush each digit from each other:

```python
def scan_rect(image, rect = None):
  p = image.load()
  eigens = []
  current = 0
  x_range = range(image.size[0])
  y_range = range(image.size[1])
  zero = p[image.size[0] - 1, image.size[1] - 1]
  if rect:
    x_range = range(rect['left'], rect['right'])
    y_range = range(rect['top'], rect['bottom'])
  for x in x_range:
    count = 0
    for y in y_range:
      if p[x, y] != zero:
        count += 1
    if count != 0:
      current = current * 10 + count
    # Void
    if current != 0 and count == 0:
      eigens.append(current)
      current = 0
  return eigens
```

```python
scan_rect(sample)
```

Guess what, we are lucky. Then it's very straight forward:

```python
def build_eigen_map():
  sample_image = Image.open('sample.bmp')
  sample_eigens = scan_rect(sample_image)
  eigen_map = {}
  for i in range(10):
    eigen_map[sample_eigens[i]] = i
  return eigen_map

E = build_eigen_map()

def recognize(E, path):
  image = Image.open(path)
  rect = RANGE_RECT.get(image.size)
  if not rect:
    print "Unknown image size %s, abort" % (image.size)
    return
  eigens = scan_rect(image, rect)
  number = 0
  for e in eigens:
    number = number * 10 + E[e]
  return number
```

Finally:

```python
print "Range: %d" % (recognize(E, 'radar1.gif'))
```
